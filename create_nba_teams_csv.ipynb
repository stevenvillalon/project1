```python
# Dependencies
import pandas as pd
from nba_api.stats.static import teams
```

```python
# Create df of NBA Teams
nba_teams = teams.get_teams()
nba_teams_df = pd.DataFrame(nba_teams)
nba_teams_df
```

```python
# Export to CSV without the index
nba_teams_df.to_csv('nba_teams.csv', index=False)
```
